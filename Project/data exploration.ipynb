```python
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('./passengers.dataframe')
df['Time'] = pd.to_datetime(df['Time'], format="%H:%M")
stations = df.columns[2:]
```

```python
print(df.shape)
df
```

### Top Source Stations Over Entire Day

```python
df.groupby('FROM').sum().sum(axis=1).sort_values(ascending=False)
```

### Top Destinations Over Entire Day

```python
df[stations].sum(axis=0).sort_values(ascending=False)
```

### Arriving passengers at time of day

```python
df.groupby('Time').sum().sum(axis=1)
```

```python
time_traffic = df.groupby('Time').sum().sum(axis=1)
plt.plot(time_traffic.values)
labels = list(df['Time'].unique());
labels = pd.to_datetime(labels).strftime('%H:%M')
labels = [label if i%int(len(labels)/20)==0 else '' for i, label in enumerate(labels) ]
plt.xticks(list(range(0,len(time_traffic))), labels, rotation='vertical');
xlabel('Time of day')
ylabel('Total Num Of Arriving Passengers')
```
